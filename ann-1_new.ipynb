```python
import math
import numpy as np
import matplotlib.pyplot as plt
# LINEAR ACTIVATION FUNCTION
def linear_func(x):
    return x
def plot_linear():
    x = np.linspace(-10, 10, 100) 
    y = x
 
    plt.plot(x, y)
    plt.xlabel('Input')
    plt.ylabel('Linear Output')
    plt.title('Linear Activation Function')
    plt.grid(True)
    plt.show()
print(linear_func(62))
plot_linear()
```

```python
# BINARY ACTIVATION FUNCTION
def binary_activation(x, threshold=0):
    return np.where(x < threshold, 0, 1)
def plot_binary():
    x = np.linspace(-5, 5, 100)
    threshold = 0
    y = binary_activation(x, threshold)
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label='Binary Activation Function')
    plt.title('Binary Activation Function')
    plt.xlabel('Input')
    plt.ylabel('Output')
    plt.grid(True)
    plt.legend()
    plt.ylim([-0.1, 1.1]) 
    plt.show()
print(binary_activation(12))
plot_binary()
```

```python
#NON-LINEAR ACTIVATION FUNCTION
#Sigmoid Acivation Function
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
def plot_sigmoid():
    x = np.linspace(-10, 10, 100) 
    y = 1 / (1 + np.exp(-x)) 
    
    plt.plot(x, y)
    plt.xlabel('Input')
    plt.ylabel('Sigmoid Output')
    plt.title('Sigmoid Activation Function')
    plt.grid(True)
    plt.show()
result = sigmoid(30)
print(result)
plot_sigmoid()
```

```python
#(tanh)
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
def plot_tanh():
    x = np.linspace(-10, 10, 100)
    tanh = np.tanh(x)
    plt.plot(x, tanh)
    plt.title("Hyperbolic Tangent (tanh) Activation Function")
    plt.xlabel("x")
    plt.ylabel("tanh(x)")
    plt.grid(True)
    plt.show()
print(tanh(56))
plot_tanh()
```

```python
# Rectified Linear Unit (ReLU) 
def relu(x):
    return max(0,x)
def plot_relu():
    x = np.linspace(-10, 10, 100)
    relu = np.maximum(0, x)
    plt.plot(x, relu)
    plt.title("ReLU Activation Function")
    plt.xlabel("x")
    plt.ylabel("ReLU(x)")
    plt.grid(True)
    plt.show()
print(relu(27))
plot_relu()
```

```python
# Leaky Rectified Linear Unit (ReLU)
def leaky_relu(x, alpha=0.1):
    return max(x, alpha * x)
def plot_leaky_relu():
    x = np.linspace(-10, 10, 100)

    def leaky_relu(x, alpha=0.1):
        return np.where(x >= 0, x, alpha * x)

    leaky_relu_values = leaky_relu(x)
    
    plt.plot(x, leaky_relu_values)
    plt.title("Leaky ReLU Activation Function")
    plt.xlabel("x")
    plt.ylabel("Leaky ReLU(x)")
    plt.grid(True)
    plt.show()
print(leaky_relu(45))
plot_leaky_relu()
```
